```python
import xml.etree.ElementTree as ET
import pandas as pd
from IPython.display import display

# parse xml file
tree = ET.parse("/content/100_Welding_Videos_Frames_RGB_labeled/3.xml")
root = tree.getroot() # get root object

bbox_coordinates = []

for member in root.findall('object'):
    #class_name = member[0].text # class name

    # bbox coordinates
    xmin = int(member[4][0].text)
    ymin = int(member[4][1].text)
    xmax = int(member[4][2].text)
    ymax = int(member[4][3].text)
    # store data in list
    bbox_coordinates.append([xmin, ymin, xmax, ymax])


print(bbox_coordinates)
print(bbox_coordinates[0][0])

df = pd.DataFrame.from_records(bbox_coordinates)
display(df)

df.to_csv(r'/content/100_Welding_Videos_Frames_RGB_labeled/3.txt', header=None, index=None, sep=' ', mode='a')
```

```python
height = int(root.find("size")[0].text)
width = int(root.find("size")[1].text)
channels = int(root.find("size")[2].text)

print(height)
print(width)
print(channels)
```
